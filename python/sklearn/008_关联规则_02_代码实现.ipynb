```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = {
    'ID':[1,2,3,4,5,6],
    'Onion':[1,0,0,1,1,1],
    'Potato':[1,1,0,1,1,1],
    'Burger':[1,1,0,0,1,1],
    'Milk':[0,1,1,1,0,1],
    'Beer':[0,0,1,0,1,0]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# frequent_itemsets = apriori(df[['Onion','Potato','Burger','Milk','Beer']].astype('bool'), min_support=0.5, use_colnames=True)
frequent_itemsets = apriori(df.drop(labels='ID',axis=1).astype('bool'), min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules
```

```python
rules[(rules['lift']>1.125) & (rules['confidence']>0.8)]
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
retail_shopping_basket = {
    'ID':[1,2,3,4,5],
    'Basket':dataset
}

retail = pd.DataFrame(retail_shopping_basket)
retail
```

```python
retail_id = retail.drop(labels='Basket',axis=1)
retail_id
```

```python
retail_Basket = retail.Basket.str.join(',')
retail_Basket
```

```python
retail_Basket = retail_Basket.str.get_dummies(',')
retail_Basket
```

```python
type(retail_Basket)
```

```python
retail = retail_id.join(retail_Basket)
retail
```

```python
frequent_itemsets_2 = apriori(retail.drop(labels='ID',axis=1).astype('bool'), min_support=0.6, use_colnames=True)
frequent_itemsets_2
```
