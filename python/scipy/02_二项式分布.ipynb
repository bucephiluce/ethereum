```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style
from IPython.core.display import HTML

# 由于 %matplotlib inline 的存在，当输入plt.plot(x,y_1)后，不必再输入 plt.show()，图像将自动显示出来
# plotting config
%matplotlib inline

style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (14,7)

plt.figure(dpi=100)

# pmf
plt.bar(np.arange(21),
        (stats.binom.pmf(np.arange(21), p=.5, n=20)),
        width=.75,
        alpha=0.75
        )

# cdf
plt.plot(np.arange(21),
        stats.binom.cdf(np.arange(21), p=.5, n=20),
        color='#fc4f30'
        )

# legend
plt.text(x=4.5, y=.7, s='pmf (normed)', alpha=.75, weight='bold', color='#008fd5')
plt.text(x=14.5, y=.9, s='cdf', alpha=.75, weight='bold', color='#008fd5')

# ticks
plt.xticks(range(21)[::2])
plt.tick_params(axis='both', which='major', labelsize=18)
plt.axhline(y=0.005, color='black',linewidth=1.3, alpha=.7)
```
