```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style
from IPython.core.display import HTML

# 由于 %matplotlib inline 的存在，当输入plt.plot(x,y_1)后，不必再输入 plt.show()，图像将自动显示出来
# plotting config
%matplotlib inline

style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (14,7)

plt.figure(dpi=100)

plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100))/np.max(stats.norm.pdf(np.linspace(-3,3,100)))
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100))/np.max(stats.norm.pdf(np.linspace(-3,3,100))),
                alpha=.15
                )
# cdf
plt.plot(np.linspace(-4,4,100),
        stats.norm.cdf(np.linspace(-4,4,100)),
        color='#fc4f30'
        )

# legend
plt.text(x=-1.5, y=.7, s='pdf (normed)',rotation=65, alpha=.75, weight='bold', color='#008fd5')
plt.text(x=-.5, y=.5, s='cdf',rotation=55, alpha=.75, weight='bold', color='#008fd5')

# ticks
plt.tick_params(axis='both', which='major', labelsize=18)
plt.axhline(y=0.0, color='black',linewidth=1.3, alpha=.7)

# title
plt.text(x=-5, y=1.25, s='Normal Distribution', alpha=.75, weight='bold',fontsize=26)
plt.text(x=-5, y=1.1, s='Depicted below ...', alpha=.85, fontsize=19)

```

```python
plt.figure(dpi=100)

# pdf mu = 0
plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100),loc=0)
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100),loc=0),
                alpha=.15
                )
# pdf mu = 2
plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100),loc=2)
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100),loc=2),
                alpha=.15
                )

# pdf mu = -2
plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100),loc=-2)
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100),loc=-2),
                alpha=.15
                )

# legend
plt.text(x=-1, y=.35, s='$ \mu = 0$',rotation=65, alpha=.75, weight='bold', color='#008fd5')
plt.text(x=1, y=.35, s='$ \mu = 2$',rotation=65, alpha=.75, weight='bold', color='#fc4f30')
plt.text(x=-3, y=.35, s='$ \mu = -2$',rotation=65, alpha=.75, weight='bold', color='#e5ae38')

# ticks
plt.tick_params(axis='both', which='major', labelsize=18)
plt.axhline(y=0.0, color='black',linewidth=1.3, alpha=.7)

# title
plt.text(x=-5, y=0.51, s='Normal Distribution - $ \mu $', alpha=.75, weight='bold',fontsize=26)
plt.text(x=-5, y=0.45, s='Depicted below ...', alpha=.85, fontsize=19)

```

```python
plt.figure(dpi=100)

# pdf sigma = 1
plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100),scale=1)
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100),scale=1),
                alpha=.15
                )
# pdf sigma = 2
plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100),scale=2)
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100),scale=2),
                alpha=.15
                )

# pdf sigma = -0.5
plt.plot(np.linspace(-4,4,100), 
        stats.norm.pdf(np.linspace(-4,4,100),scale=0.5)
        )
plt.fill_between(
                np.linspace(-4,4,100), 
                stats.norm.pdf(np.linspace(-4,4,100),scale=0.5),
                alpha=.15
                )

# legend
plt.text(x=-1.25, y=.25, s='$ \sigma = 1$',rotation=65, alpha=.75, weight='bold', color='#008fd5')
plt.text(x=-2., y=.15, s='$ \sigma = 2$',rotation=65, alpha=.75, weight='bold', color='#fc4f30')
plt.text(x=-0.65, y=.65, s='$ \sigma = 0.5$',rotation=65, alpha=.75, weight='bold', color='#e5ae38')

# ticks
plt.tick_params(axis='both', which='major', labelsize=18)
plt.axhline(y=0.0, color='black',linewidth=1.3, alpha=.7)
# plt.axvline(x=-4.4, color='black',linewidth=1.3, alpha=.7)

# title
plt.text(x=-5, y=0.91, s='Normal Distribution - $ \sigma $', alpha=.75, weight='bold',fontsize=26)
plt.text(x=-5, y=0.85, s='Depicted below ...', alpha=.85, fontsize=19)

```

## 随机的几个样本
可以使用norm.rvs()其中默认$\mu=0$, $\sigma=1 $; 也可以自己指定

```python
from scipy.stats import norm
print(norm.rvs(), end='\n\n')
print(norm.rvs(size=10), end='\n\n')
print(norm.rvs(loc=10,scale=0.1), end='\n\n')
print(norm.rvs(loc=10,scale=0.1,size=10), end='\n\n')
```

## 概率密度函数(Probability Density Function)
根据我们的密度分布函数来计算

```python
from scipy.stats import norm

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (14,7)

x = -1
y = 3
print('pdf(x) = {}\npdf(y)={}'.format(norm.pdf(x),norm.pdf(y)))
print('ppf(0.01) = {}\nppf(0.99)={}'.format(norm.ppf(0.01),norm.ppf(0.99)))

# x_s = np.linspace(-3,3,50)
x_s = np.linspace(norm.ppf(0.01),norm.ppf(0.99),100)
y_s = norm.pdf(x_s)
plt.scatter(x_s,y_s)
```

## 累积概率密度函数(Cumulative Probability Density Function)

```python
from scipy.stats import norm

print('P(X<0.3)={}'.format(norm.cdf(0.3)))
print('P(-0.2<X<0.2)={}'.format(norm.cdf(0.2) - norm.cdf(-0.2)))
```

## 基于数据画分布

```python
plt.figure(dpi=100)

mu_real = 10
sigma_real = 2

# 产生1000个样本
np.random.seed(42)
sample = stats.norm.rvs(loc=mu_real, scale=sigma_real,size=1000)

# 估计期望和方差
mu_est = np.mean(sample)
sigma_est = np.std(sample)

print('Estimated MU: {} \nEstimated SIGMA: {}'.format(mu_est, sigma_est))

# sample distribution density=True归一化操作
plt.hist(sample, bins=50, alpha=.25 ,density=True)

# True
plt.plot(np.linspace(2,18,1000), 
        stats.norm.pdf(np.linspace(2,18,1000),scale=sigma_real,loc=mu_real)
        )
# Estimated
plt.plot(np.linspace(2,18,1000), 
        stats.norm.pdf(np.linspace(2,18,1000),scale=sigma_est,loc=mu_est)
        )

# legend
plt.text(x=9.5, y=.1, s='sample', alpha=.75, weight='bold', color='#008fd5')
plt.text(x=7., y=.2, s='true distribution',rotation=55, alpha=.75, weight='bold', color='#fc4f30')
plt.text(x=5, y=.12, s='estimated distribution',rotation=55, alpha=.75, weight='bold', color='#e5ae38')

# ticks
plt.tick_params(axis='both', which='major', labelsize=18)
plt.axhline(y=0.0, color='black',linewidth=1.3, alpha=.7)
# plt.axvline(x=-4.4, color='black',linewidth=1.3, alpha=.7)

# title
plt.text(x=0, y=0.3, s='Normal Distribution', alpha=.75, weight='bold',fontsize=26)
```
