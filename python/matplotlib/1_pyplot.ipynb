```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('天气数据.xlsx')

#用于处理中文
plt.rcParams['font.sans-serif']=['SimHei']
#设置画布的颜色
plt.figure(facecolor='yellow',figsize=(15,5))

x = df['日期']
y = df['温度']

plt.xlabel('日期') #对x轴添加标题
plt.ylabel('室外温度')#对y轴添加标题
plt.plot(x,y ,marker='o',color='r' , mfc='w') #markerfacecolor简写mfc
plt.show()
```

```python
# 设置坐标轴的刻度
import matplotlib.pyplot as plt
import random

x = [i for i in range(1, 11)]
y = [random.randint(1, 10) for _ in range(10)]

plt.plot(x, y, marker='o', mfc='w')

month = [str(i)+'月' for i in range(1, 11)]

plt.xticks(range(1, 11), month)  # 设置坐标轴的刻度
plt.yticks(range(1, 11))  # 设置坐标轴的刻度
plt.xlim(1, 10)  # 设置坐标轴的范围

# plt.grid(color='r', linestyle='--', linewidth='1')
plt.grid(color='0.6', linestyle='--', linewidth='1', axis='y') #设置网格线 , 并且隐藏y轴网格线

plt.show()
```

```python
# 添加标签
import matplotlib.pyplot as plt
import random

x = [i for i in range(1, 11)]
y = [random.randint(1, 10) for _ in range(10)]

plt.plot(x, y, marker='o', mfc='w')

month = [str(i)+'月' for i in range(1, 11)]

plt.xticks(range(1, 11), month)  # 设置坐标轴的刻度
plt.yticks(range(1, 11))  # 设置坐标轴的刻度

#添加文斌标签
for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=12,color='r')

#添加图表的标题
plt.title('测试练习折线图' , fontsize=18)

# 添加图例
plt.legend(('销售图例',))
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[11,23,45,35,8,33]
#用于处理中文
plt.rcParams['font.sans-serif']=['SimHei']

plt.plot(x,y,marker='o')

for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom' ,fontsize=12,color='r')

#添加文本注释
plt.annotate('最大值',xy=(3,45),xytext=(4,45),arrowprops=dict(facecolor='r',shrink=0.05))
plt.show()
```

```python
#图表与画布边缘间距
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[11,23,45,35,8,33]
#用于处理中文
plt.rcParams['font.sans-serif']=['SimHei']

plt.plot(x,y,marker='o')

for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom' ,fontsize=12,color='r')

#添加文本注释
plt.annotate('最大值',xy=(3,45),xytext=(4,45),arrowprops=dict(facecolor='r',shrink=0.05))


#设置间距
plt.subplots_adjust(left=0.2,right=0.9,top=0.9,bottom=0.2)
plt.show()
```
