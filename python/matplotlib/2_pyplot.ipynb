```python
# 折线图
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_excel('学生成绩表.xlsx')

x = df['姓名']
y1 = df['语文']
y2 = df['数学']
y3 = df['英语']

# 进行分组与求和
# df.groupby('日期')['语文'].sum()
# 总分可以这样相加
# y = df['语文'] + df['数学'] + df['英语']

# 解决中文乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei']

# 设置画布大小
plt.figure(figsize=(13, 6))

plt.plot(x, y1, color='r', marker='o', mfc='w', label='语文')
plt.plot(x, y2, color='g', marker='o', mfc='w', label='数学', linestyle='--')
plt.plot(x, y3, color='b', marker='o', mfc='w', label='英语', linestyle='-.')

# 设置网格线
plt.grid(axis='y')

# 设置坐标轴的标题
plt.xlabel('姓名')
plt.ylabel('分数')

# 设置图表的图例
plt.legend(['语文', '数学', '英语'])

# 设置坐标轴刻度
plt.yticks(range(0, 101, 10))

plt.show()
```

```python
#绘制简单的柱状图
import matplotlib.pyplot as plt
import random

x = [1,2,3,4,5]
height = [random.randint(10000,1000000) for i in range(5)]

# 解决中文乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(13,6))

plt.grid(axis='y' , linestyle='--')
plt.xlabel('产品名称')
plt.ylabel('销售总量')
#设置文本标签
for a,b in zip(x,height):
    plt.text(a, b, format(b,','), ha='center', va='bottom', color='r', fontsize=12,alpha=0.9)

plt.bar(x, height, width=0.5,alpha=0.5)

plt.legend(('销售额',))

plt.show()
```

```python
#绘制简单的饼形图
import matplotlib.pyplot as plt

x = [10,34,5,56,58,65]

plt.pie(x,autopct='%1.1f%%',startangle=90,textprops={'fontsize':12},
        explode=(0,0,0,1,0,1),shadow=True)

plt.axis('equal')
plt.legend(x , loc='upper right' , frameon=False)
plt.show()
```

```python
#绘制简单的饼形图
import matplotlib.pyplot as plt

x = [10,34,5,56,58,65]
x2 = [35,45,70,20,40,61]
plt.pie(x,autopct='%1.1f%%',startangle=90,pctdistance=0.8,textprops={'fontsize':12},radius=1,
        wedgeprops={'width':0.4,'edgecolor':'w'})
plt.pie(x2,autopct='%1.1f%%',startangle=90,pctdistance=0.7,textprops={'fontsize':12},radius=0.7,
        wedgeprops={'width':0.4,'edgecolor':'w'})
plt.axis('equal')
plt.legend(x , loc='upper right' , frameon=False)
plt.show()
```

```python
# 折线图
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_excel('学生成绩表.xlsx')

# 解决中文乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei']

x = df.loc[:, '语文':'英语'].values

# 设置x轴的刻度
plt.xticks(range(3), ['语文', '数学', '英语'])
plt.yticks(range(10), df['姓名'])

plt.imshow(x)
# 显示颜色条
plt.colorbar()

plt.title('学生成绩热力图', fontsize=18)

plt.show()
```

```python
# 绘制多箱型图
import matplotlib.pyplot as plt

x = [17, 3, 54, 7, 9]
plt.boxplot(
    x,
    patch_artist=True, # 填充箱子颜色
    showmeans=True, # 显示均值
    flierprops={'mfc':'r','mec':'b','markersize':10}, # mec=markeredgecolor , 设置异常值的填充色,边框颜色,大小
    meanprops={'marker':'v'}, # 设置均值的符号,颜色,大小
    medianprops={'linestyle':'--'}
)
plt.show()
```

```python
#绘制子图
import matplotlib.pyplot as plt
import random

# plt.subplot(2,2,1) # 第一个绘图区,2行2列,第一个图
plt.subplot(221) # 第一个绘图区,2行2列,第一个图 上一行的简写
plt.plot([1,2,3,4,5],[random.randint(10,20) for i in range(5)])

plt.subplot(2,2,2) # 第二个绘图区,2行2列,第2个图
plt.plot([1,2,3,4,5],[random.randint(10,20) for i in range(5)],'ro')

plt.subplot(2,1,2) # 第二个绘图区,2行1列,第3个图
plt.bar([1,2,3,4,5],[random.randint(10,20) for i in range(5)])

plt.show()
```

```python
# 绘制子图的第二种方式
import matplotlib.pyplot as plt
import random

figure, axes = plt.subplots(2, 2)

axes[0,0].plot([1,2,3,4,5],[random.randint(10,20) for i in range(5)])
axes[0,1].plot([1,2,3,4,5],[random.randint(10,20) for i in range(5)] ,'ro')
axes[1,0].bar([1,2,3,4,5],[random.randint(10,20) for i in range(5)])
axes[1,1].pie([random.randint(10,20) for i in range(5)],autopct='%1.1f%%')

plt.show()
```

```python
#绘制子图的第三种方式
import matplotlib.pyplot as plt
import random

fig = plt.figure() #绘制画布
ax1 = fig.add_subplot(2,2,1)
ax1.plot([1,2,3,4,5],[random.randint(10,20) for i in range(5)])

ax2 = fig.add_subplot(2,2,2)
ax2.plot([1,2,3,4,5],[random.randint(10,20) for i in range(5)],'ro')

ax3 = fig.add_subplot(2,2,3)
ax3.bar([1,2,3,4,5],[random.randint(10,20) for i in range(5)])

ax4 = fig.add_subplot(2,2,4)
ax4.pie([random.randint(10,20) for i in range(5)],autopct='%1.1f%%')

plt.show()
```
