```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 设置画布样式
sns.set_style('darkgrid')

# 设置中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']

# 读取数据
df = pd.read_excel('学生成绩表.xlsx')

sns.relplot(x='姓名', y='数学', kind='line', data=df)
sns.lineplot(x='姓名', y='语文', data=df , color='r')

plt.xticks(df['姓名'], rotation=60)  # 将刻度进行竖直方向刻画

plt.show()
```

```python
#绘制线性回归图
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 设置画布样式
sns.set_style('darkgrid')

# 设置中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']

# 读取数据
df = sns.load_dataset('tips') #关于小费的,是seanborn自带的

sns.lmplot(x='total_bill',y='tip' ,data=df)

plt.show()
```

```python
#箱型图
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 设置画布样式
sns.set_style('darkgrid')

# 设置中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']

# 读取数据
df = sns.load_dataset('tips') #关于小费的,是seanborn自带的

sns.boxplot(x='day',y='total_bill',hue='time',data=df)

plt.show()
```

```python
#绘制核密度图
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 设置画布样式
sns.set_style('darkgrid')

# 设置中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']

# 读取数据
df = sns.load_dataset('iris') #关于小费的,是seanborn自带的

sns.kdeplot(df['petal_width'], shade=True, color='r')
sns.kdeplot(df['petal_length'], shade=True, color='b')

plt.show()
```

```python
#绘制边际核密度图
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 设置画布样式
sns.set_style('darkgrid')

# 设置中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']

# 读取数据
df = sns.load_dataset('iris') #关于小费的,是seanborn自带的

sns.jointplot(x=df['sepal_length'],y=df['sepal_width'],kind='kde',space=0)

plt.show()
```
